INTRODUCTION: Decision Tree is a Supervised learning technique that can be used for both classification and Regression problems, 
but mostly it is preferred for solving Classification problems.
It is a tree-structured classifier, where internal nodes represent the features of a dataset,
branches represent the decision rules and each leaf node represents the outcome.

```python
#step 1; Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split

```

```python
# step 2:Load the iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head() # See the first 5 rows
```

```python
x=iris.data
y=iris.target
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#step 3:Model building
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(max_depth=4)
dtree.fit(X_train,y_train)
y_pred=dtree.predict(X_test)
```

```python
#visualizing the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize=(10,10)) # set plot size (denoted in inches)
plot_tree(dtree, fontsize=10)
plt.show()
```

```python
#Evaluation of the model
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of the model is :",accuracy_score(y_test,y_pred))

print("Confusion matrix is:",confusion_matrix(y_test,y_pred))
```

```python
#Predict unkown data with custum input
cust=dtree.predict(np.array([12.3,43.2,12.2,11.0]).reshape(1,-1))
print(cust)
print(iris.target_names)
```
